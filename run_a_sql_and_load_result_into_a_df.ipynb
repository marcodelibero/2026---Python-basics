```python
query = """
SELECT *
FROM `leo-production-bi.DW_Fact_Management_Analysis.fact_core_daily_no_personal_history`
WHERE date = CURRENT_DATE - 1
LIMIT 10
"""

project_id = 'leo-production-bi' # Use the project ID where the table resides

df = pandas_gbq.read_gbq(
    query,
    project_id=project_id,
    progress_bar_type='tqdm',
    # reauth=True, # reauth is not needed when using auth.authenticate_user()
    # credentials=credentials, # credentials are not needed when using auth.authenticate_user()
)

display(df.head())
```
