# 🔎 Basic Filtering

In this section we are going to introduce some basic command to filter a dataframe over multiple condition

```python
filtered_tips_df = tips_df[(tips_df['sex'] == 'Female') & (tips_df['size'] == 2)]
display(filtered_tips_df.head(10))
```

### Filtering with an 'OR' Condition

Let's filter for rows where the `smoker` is 'Yes' **OR** the `day` is 'Sun'.

```python
filtered_by_or = tips_df[(tips_df['smoker'] == 'Yes') | (tips_df['day'] == 'Sun')]
display(filtered_by_or.head())
```

### Filtering using `isin()` for multiple values

Sometimes you want to filter for rows where a column's value is in a list of possible values. Let's find all tips that occurred on 'Thur' or 'Fri'.

```python
filtered_by_isin = tips_df[tips_df['day'].isin(['Thur', 'Fri'])]
display(filtered_by_isin.head())
```

### Filtering for Numerical Ranges

You can also filter numerical columns for values within a specific range. Let's find all entries where the `total_bill` is between 20 and 30 (inclusive).

```python
filtered_by_range = tips_df[(tips_df['total_bill'] >= 20) & (tips_df['total_bill'] <= 30)]
display(filtered_by_range.head())
```

# **🧪 Statistical Tests**
The statistical test choice depends on the **hypothesis** you would like to test and the **data**. Here we will demonstate a test specific to the hypothesis we would like to test and the data we have.

```python
# Does tip amount differ by whether customer is a smoker or not ?
```

```python
tips_df.head()
```

```python
# Here we have unpaired data
```

```python
#testing for normality
from scipy.stats import kstest, norm
#seperate the data based on whether customer is a smoker or not
smoker_data=tips_df[tips_df['smoker']=='Yes']['tip']
non_smoker_data=tips_df[tips_df['smoker']=='No']['tip']
#standarizing the data for KS test
smoker_std=(smoker_data-smoker_data.mean())/smoker_data.std()
non_smoker_std=(non_smoker_data-non_smoker_data.mean())/non_smoker_data.std()
#KS test
smoker_ks_stat, smoker_p = kstest(smoker_std, 'norm')
nonsmoker_ks_stat, nonsmoker_p = kstest(non_smoker_std, 'norm')

print(f"Smoker KS p-value: {smoker_p:.4f}")
print(f"Non-Smoker KS p-value: {nonsmoker_p:.4f}")
```

```python
tips_df['tip'].min()
```

```python
# we can also see normality by plotting distributions
import matplotlib.pyplot as plt
sns.histplot(smoker_std, kde=True, label='Smoker',color='#c11717')
sns.histplot(non_smoker_std, kde=True, label='Non-Smoker',color='#d9b608')
plt.legend()
plt.xlabel("Standardized Tip")
plt.ylabel("Count")
plt.title("Tip Distribution: Smokers vs Non-Smokers")

plt.show()
```

```python
smoker_data
```

```python
#Data seems to be approx. normal as p>0.05
# We will use an independent samples t-test
from scipy.stats import ttest_ind
stat, p_val = ttest_ind(smoker_data, non_smoker_data)
print(f"statistic={stat}, p_val={p_val}")
```

```python
#pvalue>0.05.-> No statistically significant difference between tips from smokers and non-smokers
```

```python
# 0. Loading the bigquery translator
!pip install sqlalchemy-bigquery --quiet

import pandas as pd
from sqlalchemy import create_engine
from google.colab import auth

# 1. Google autenthication
auth.authenticate_user()

# 2. creating the bigquery engine
engine = create_engine('bigquery://leo-production-bi')
```
