```python
import pandas as pd
import seaborn as sns

# Load the data

tips_df = sns.load_dataset("tips")
#add a column with date
tips_df['date']=pd.date_range(start='2025-01-01', periods=len(tips_df), freq='D')
```

# **↩ Changing Column Names**

```python
#checking all column names
tips_df.columns
```

```python
#change column name size to be more intuitive
tips_df.rename(columns={'size':'party_size'},inplace=True)
```

```python
tips_df
```

# **↩ Changing data types**




```python
#check distinct values in size
tips_df['size'].unique()
```

```python
#changing size to str
tips_df['size']=tips_df['size'].astype('category')
tips_df.dtypes
```

```python
tips_df.isnull().sum()
```

# **⏲** **Changing daily data to monthly/yearly**




```python
# what are the monthly sales ?
#create a new column with year-month
tips_df['year_month']=tips_df['date'].dt.to_period('M')
tips_df.sort_values(by='date',inplace=True)
tips_df
```

```python
#monthly sales
monthly_sales=tips_df.groupby('year_month')['total_bill'].sum()
monthly_sales
```

```python
monthly_sales.sum()
```

```python
tips_df['year']=tips_df['date'].dt.year
yearly_sales=tips_df.groupby('year')['total_bill'].sum()
yearly_sales
```

# **⚠ Null Values**


```python
#setting tip values >9 to null
tips_df.loc[tips_df['tip']>9,'tip']=None
```

```python
#fidning null counrts in each column
tips_df.isnull().sum()
```

```python
#replacing null value with 10
tips_df['tip']=tips_df['tip'].fillna(10)
```

```python
#check to see null value replaced
tips_df.isnull().sum()
```

# ⚖ Group By

```python
#find the sales generated by time of service
tips_df.groupby('time')['total_bill'].sum()
```
