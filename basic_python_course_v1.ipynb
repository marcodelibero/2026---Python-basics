# **⏲** **Changing daily data to monthly/yearly**




```python
# what are the monthly sales ?
#create a new column with year-month
tips_df['year_month']=tips_df['date'].dt.to_period('M')
tips_df.sort_values(by='date',inplace=True)
tips_df
```

```python
#monthly sales
monthly_sales=tips_df.groupby('year_month')['total_bill'].sum()
monthly_sales
```

```python
monthly_sales.sum()
```

```python
tips_df['year']=tips_df['date'].dt.year
yearly_sales=tips_df.groupby('year')['total_bill'].sum()
yearly_sales
```

# **⚠ Null Values**


```python
#setting tip to nulll when tip is >9
tips_df.loc[tips_df['tip']>9,'tip']=None
```

```python
#fidning null counrts in each column
tips_df.isnull().sum()
```

```python
#replacing null value with 10
tips_df['tip']=tips_df['tip'].fillna(10)
```

```python
#check to see null value replaced
tips_df.isnull().sum()
```

# ➕ ➖ **Basic Aritmethic Operations**

What is the % of tip offered from the total bill ?


```python
tips_df['%tip_from_total_bill']=(tips_df['tip']/tips_df['total_bill'])*100
tips_df['%tip_from_total_bill']=tips_df['%tip_from_total_bill'].round(1)
tips_df.head()
```

# **🏷 Labelling Data**
Let's say you would like to flag sales occuring on Public Holidays

```python
#let's say we have the days of holidays in a list

public_holidays=['2025-01-01','2025-02-10','2025-03-19','2025-03-31']
```

```python
#creating a column that labels sales on public holidays
tips_df['public_holiday']=tips_df['date'].isin(public_holidays)
```

```python
tips_df.head()
```

Labelling based on Conditions

```python
tips_df.shape
```

```python

# Define the logical conditions for selection
# You can combine multiple criteria using bitwise operators: & (AND), | (OR)
conditions = [
    (tips_df['%tip_from_total_bill'] <10) ,
    (tips_df['%tip_from_total_bill'] >=10)
]

# Define the corresponding values to assign for each condition
choices = [ 'Silver', 'Gold']

# Create the new column using np.select
# The 'default' argument specifies the value to use if no conditions are met
tips_df['customer_tier'] = np.select(conditions, choices,'Unclassified')

tips_df.head(5)
```

# ⚖ Group By

```python
#find the sum of  sales generated by time of service
tips_df.groupby('time')['total_bill'].sum()
```

```python
# Applying group by with agg
#find the sum of sales and mean tip generated by time of service
tips_df.groupby('time').agg({'total_bill':'sum','tip':'mean' }).rename(columns={'total_bill':'total_bill_sum','tip':'tip_mean'})
#
```

# **➿ For Loops**

```python
tips_df.head()
```

```python
tips_df.groupby('day')
```

```python
#for each day of the week we find the mean sales and mean tip
for day, group in tips_df.groupby('day'):
  avg_tip=group['tip'].mean()
  avg_sales=group['total_bill'].mean()
  print(f"{day}:avg_tip,avg_sales:{avg_tip},{avg_sales}")
```

```python
#this can also be achieved with a group by
tips_df.groupby('day').agg({'tip':'mean','total_bill':'mean'})
```
